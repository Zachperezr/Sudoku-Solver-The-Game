```python
#creating template for baord
board = [
        [7, 8, 0, 4, 0, 0, 1, 2, 0],
        [6, 0, 0, 0, 7, 5, 0, 0, 9],
        [0, 0, 0, 6, 0, 1, 0, 7, 8],
        [0, 0, 7, 0, 4, 0, 2, 6, 0],
        [0, 0, 1, 0, 5, 0, 9, 3, 0],
        [9, 0, 4, 0, 6, 0, 0, 0, 5],
        [0, 7, 0, 3, 0, 0, 0, 1, 2],
        [1, 2, 0, 0, 0, 7, 4, 0, 0],
        [0, 4, 9, 2, 0, 6, 0, 0, 7]
    ]

#creating recursion algorithim
def solve(bo):

    find = find_empty(bo)
    if not find:
        return True
    else:
        row, col = find
    
    for i in range(1,10):
        if valid(bo, i, (row, col)):
            bo[row][col] = i

            if solve(bo):
                return True
            
            bo[row][col] = 0
   
    return False

#checking validity of sudoku puzzle
def valid(bo, num, pos):
     
    #check row
    for i in range(len(bo[0])):
        if bo[pos[0][i]] == num and pos[1] != i:
            return False
    
    #check column
    for i in range(len(bo)):
        if bo[i][pos[1]] == num and pos[0] != i:
            return False
          
    #check box
    box_x = pos[1] // 3
    box_y = pos[0] // 3

    for i in range(box_y * 3, box_y * 3 + 3):
        for j in range(box_x * 3, box_x * 3 + 3):
            if bo[i][j] == num and (i,j) != pos:
                return False
    


#creating a demo board
def print_board(bo):
    for i in range(len(bo)):
        if i % 3 == 0 and i != 0:
            print("--------------------------")

        for j in range(len(bo[0])):
            if j % 3 == 0 and j!= 0:
                print(" | ", end="")

            if j == 8:
                    print(bo[i][j])
            else:
                print(str(bo[i][j]) + " ", end="")

#function to find empty spaces in sudoku puzzle
def find_empty(bo):
     for i in range (len(bo)):
          for j in range(len(bo[0])): #using "0" to identify empty spaces
               if bo[i][j] == 0:
                   return (i, j) # row, col
     return None

solve(board)
```
